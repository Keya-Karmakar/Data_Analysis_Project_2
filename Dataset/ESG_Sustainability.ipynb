```python
import numpy as n
import pandas as p
n.random.seed(42)
num=40000
industries=["Technology","Manufacturing","Retail","Healthcare","Finance","Energy","Telecom"]
data=p.DataFrame({
    "Company_ID":n.arange(100001,100001+num),
    "Industry":n.random.choice(industries,num),
    "Revenue_Mission":n.random.uniform(10,10000,num),
    "Carbon_Emission_Tons":n.random.uniform(50,50000,num),
    "Renewable_Energy_Percent":n.random.uniform(0,100,num),
    "Employee_Diversity_Score":n.random.uniform(40,100,num),
    "Safety_Incidents":n.random.randint(0,50,num),
    "Governance_Score":n.random.uniform(50,100,num),
    "Employee_Satisfaction":n.random.uniform(40,100,num),
    "Training_Hours":n.random.uniform(5,150,num)
})
data.to_csv("ECG_Sustainability_csv.csv",index=False)
```
